```python
# Solución — Ejercicio 1: Crear y explorar arrays
# Qué hace: crea los 5 arrays pedidos y muestra sus propiedades
# Para qué sirve: practicar la creación y exploración de arrays antes de usarlos con datos reales

import numpy as np

# a) Array desde lista
arr_a = np.array([5, 10, 15, 20, 25, 30])
print("a) Array desde lista:", arr_a)
print(f"   dtype={arr_a.dtype}, shape={arr_a.shape}, ndim={arr_a.ndim}, size={arr_a.size}")

# b) Array de 8 ceros
arr_b = np.zeros(8)
print("\nb) Array de ceros:", arr_b)
print(f"   dtype={arr_b.dtype}, shape={arr_b.shape}, ndim={arr_b.ndim}, size={arr_b.size}")

# c) Pares del 0 al 20 con arange
arr_c = np.arange(0, 21, 2)
print("\nc) Pares 0-20:", arr_c)
print(f"   dtype={arr_c.dtype}, shape={arr_c.shape}, ndim={arr_c.ndim}, size={arr_c.size}")

# d) 6 valores igualmente espaciados entre 0 y 100
arr_d = np.linspace(0, 100, 6)
print("\nd) linspace(0, 100, 6):", arr_d)
print(f"   dtype={arr_d.dtype}, shape={arr_d.shape}, ndim={arr_d.ndim}, size={arr_d.size}")

# e) Matriz 3×3 con valores del 1 al 9
arr_e = np.arange(1, 10).reshape(3, 3)
print("\ne) Matriz 3×3:")
print(arr_e)
print(f"   dtype={arr_e.dtype}, shape={arr_e.shape}, ndim={arr_e.ndim}, size={arr_e.size}")

# Reflexión: arr_a es int64 porque los elementos son enteros
# arr_b es float64 porque np.zeros() por defecto crea flotantes
print("\nReflexión: np.zeros() crea float64 por defecto; np.array([5,10,...]) infiere int64 de enteros Python")
```

```python
# Solución — Ejercicio 2: Operaciones vectorizadas con datos de ventas
# Qué hace: extrae columnas, calcula totales y aplica filtros vectorizados
# Para qué sirve: practicar el flujo completo de extracción y cálculo sobre datos reales

import pandas as pd
import numpy as np

df = pd.read_csv("../../datasets/ventas_tienda.csv")

# .values extrae los valores como array NumPy
precios = df["precio_unitario"].values
unidades = df["unidades"].values

# a) Total de venta sin bucle
totales = precios * unidades
print("a) Primeros 5 totales:", totales[:5])
print(f"   Total de registros calculados: {len(totales)}")

# b) Descuento del 10% en todos los precios
precios_con_descuento = precios * 0.90
print("\nb) Primeros 5 precios con 10% descuento:", np.round(precios_con_descuento[:5], 2))

# c) Booleano: registros con más de 10 unidades
muchas_unidades = unidades > 10
print(f"\nc) Registros con unidades > 10: {np.sum(muchas_unidades)} de {len(unidades)}")

# d) Extraer precios de los registros con muchas unidades
precios_alto_volumen = precios[muchas_unidades]
print(f"\nd) Precio promedio en registros de alto volumen: {np.mean(precios_alto_volumen):.2f}")
```

```python
# Solución — Ejercicio 3: Estadísticas sobre el array de precios
# Qué hace: calcula todas las estadísticas pedidas y el porcentaje sobre el promedio
# Para qué sirve: dominar las funciones estadísticas de NumPy sobre datos reales

import pandas as pd
import numpy as np

df = pd.read_csv("../../datasets/ventas_tienda.csv")
precios = df["precio_unitario"].values

promedio = np.mean(precios)

print(f"a) Precio promedio:          {promedio:.2f}")
print(f"b) Precio mediano:           {np.median(precios):.2f}")
print(f"c) Desviación estándar:      {np.std(precios):.2f}")
print(f"d) Precio mínimo:            {np.min(precios):.2f}")
print(f"d) Precio máximo:            {np.max(precios):.2f}")

# e) Cuántos sobre el promedio
sobre_promedio = np.sum(precios > promedio)
porcentaje = (sobre_promedio / len(precios)) * 100
print(f"\ne) Precios sobre el promedio: {sobre_promedio} registros ({porcentaje:.1f}%)")
```

```python
# Solución — Ejercicio 4: Reshaping y arrays 2D
# Qué hace: crea, reorganiza e indexa arrays 2D
# Para qué sirve: practicar el manejo de dimensiones, necesario para trabajar con matrices en ML

import numpy as np

# a) Array del 1 al 12
arr = np.arange(1, 13)
print("a) Array original:", arr)

# b) Reorganizar en 3×4
m3x4 = arr.reshape(3, 4)
print("\nb) Matriz 3×4:")
print(m3x4)

# c) Reorganizar en 4×3
m4x3 = arr.reshape(4, 3)
print("\nc) Matriz 4×3:")
print(m4x3)

# d) Segunda fila (índice 1)
print("\nd) Segunda fila de la matriz 4×3:", m4x3[1, :])

# e) Primera columna
print("e) Primera columna de la matriz 4×3:", m4x3[:, 0])

# f) Elemento fila 2, columna 1 (índice 0-based)
print(f"f) Elemento [2, 1]: {m4x3[2, 1]}")
```

```python
# Solución — Ejercicio 5 (desafío): Normalización min-max
# Qué hace: normaliza el array de precios para que todos los valores queden entre 0 y 1
# Para qué sirve: preparar datos para algoritmos de ML que son sensibles a la escala de las variables

import pandas as pd
import numpy as np

df = pd.read_csv("../../datasets/ventas_tienda.csv")
precios = df["precio_unitario"].values

# Fórmula min-max: (valor - min) / (max - min)
# Esto hace que el mínimo se convierta en 0 y el máximo en 1
precio_min = np.min(precios)
precio_max = np.max(precios)

precios_normalizados = (precios - precio_min) / (precio_max - precio_min)

print(f"Precio original — min: {precio_min:.2f}, max: {precio_max:.2f}")
print(f"Normalizado     — min: {np.min(precios_normalizados):.4f}, max: {np.max(precios_normalizados):.4f}")
print(f"\nPrimeros 5 valores normalizados: {np.round(precios_normalizados[:5], 4)}")
```
