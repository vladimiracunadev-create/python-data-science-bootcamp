```python
# Configuración inicial
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

sns.set_theme(style="whitegrid", palette="muted")

ventas = pd.read_csv("../../datasets/ventas_tienda.csv")
estudiantes = pd.read_csv("../../datasets/estudiantes.csv")
ventas["fecha"] = pd.to_datetime(ventas["fecha"])

print("Datos cargados correctamente.")
print(f"Ventas: {ventas.shape} | Estudiantes: {estudiantes.shape}")
```

```python
# SOLUCIÓN — Ejercicio 1: Distribución de ventas
# Pregunta: ¿Cómo se distribuyen las ventas diarias? ¿Hay simetría o cola?

plt.figure(figsize=(9, 4))
sns.histplot(data=ventas, x="total_venta", bins=20, kde=True, color="teal")
plt.title("Distribución de ventas diarias")
plt.xlabel("Total de venta ($)")
plt.ylabel("Frecuencia")
plt.tight_layout()
plt.show()

# Análisis descriptivo
print(f"Media:    {ventas['total_venta'].mean():.2f}")
print(f"Mediana:  {ventas['total_venta'].median():.2f}")
print(f"Asimetría (skew): {ventas['total_venta'].skew():.2f}")
print()
# Si media > mediana y skew > 0: distribución con cola hacia la derecha (hay ventas muy grandes)
# Si skew ≈ 0: distribución aproximadamente simétrica
```

```python
# SOLUCIÓN — Ejercicio 2 y 3: Boxplot vs Violinplot por categoría

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Boxplot: mediana y outliers claramente visibles
sns.boxplot(data=ventas, x="categoria", y="total_venta",
            palette="Set3", ax=axes[0])
axes[0].set_title("Ventas por categoría — Boxplot")
axes[0].set_xlabel("Categoría")
axes[0].set_ylabel("Total de venta")
axes[0].tick_params(axis="x", rotation=30)

# Violinplot: muestra la forma completa — bimodalidad, asimetría, etc.
sns.violinplot(data=ventas, x="categoria", y="total_venta",
               palette="Set3", inner="quartile", ax=axes[1])
axes[1].set_title("Ventas por categoría — Violinplot")
axes[1].set_xlabel("Categoría")
axes[1].set_ylabel("Total de venta")
axes[1].tick_params(axis="x", rotation=30)

plt.tight_layout()
plt.show()

# El violinplot revela si la distribución tiene dos picos (dos grupos mezclados)
# El boxplot solo muestra percentiles — no ves la forma completa
```

```python
# SOLUCIÓN — Ejercicio 4: Barplot por día de la semana

ventas["dia_num"] = ventas["fecha"].dt.dayofweek
nombres = {0: "Lunes", 1: "Martes", 2: "Miércoles",
           3: "Jueves", 4: "Viernes", 5: "Sábado", 6: "Domingo"}
ventas["dia_nombre"] = ventas["dia_num"].map(nombres)

orden = ["Lunes", "Martes", "Miércoles", "Jueves", "Viernes", "Sábado", "Domingo"]

plt.figure(figsize=(9, 4))
sns.barplot(data=ventas, x="dia_nombre", y="total_venta",
            order=orden, palette="Blues_d", errorbar="ci")
plt.title("Venta promedio por día de la semana")
plt.xlabel("Día")
plt.ylabel("Venta promedio ($)")
plt.tight_layout()
plt.show()

print(ventas.groupby("dia_nombre")["total_venta"].mean().reindex(orden).round(2))
```

```python
# SOLUCIÓN — Ejercicio 5 y 6: Scatterplot con hue + Heatmap

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Scatterplot
sns.scatterplot(data=ventas, x="unidades", y="total_venta",
                hue="categoria", alpha=0.6, ax=axes[0])
axes[0].set_title("Unidades vs Venta por categoría")
axes[0].legend(title="Categoría", bbox_to_anchor=(1.05, 1), loc="upper left")

# Heatmap de correlación
corr = ventas.select_dtypes(include="number").corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm",
            center=0, square=True, linewidths=0.5, ax=axes[1])
axes[1].set_title("Correlación entre variables numéricas")

plt.tight_layout()
plt.show()
```

```python
# SOLUCIÓN — Ejercicio 7: Pairplot de estudiantes

cols = ["nota_final", "asistencia", "horas_estudio", "modalidad"]
df_pair = estudiantes[cols].dropna()

g = sns.pairplot(df_pair, hue="modalidad",
                 plot_kws={"alpha": 0.45},
                 diag_kind="kde", corner=False)
g.figure.suptitle("Relaciones entre variables de estudiantes por modalidad", y=1.02)
plt.show()

# Interpretación esperada:
# - Si los puntos de nota_final vs asistencia suben de izquierda a derecha: correlación positiva
# - Si las nubes de puntos se superponen por modalidad: la modalidad no explica mucho
# - Si están separadas: la modalidad sí tiene efecto
```
