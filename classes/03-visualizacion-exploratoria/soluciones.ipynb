# Clase 03 — Visualización exploratoria

Construiremos gráficos simples para apoyar interpretación de hallazgos.

```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

base = Path.cwd()
while base.name != 'python-data-science-bootcamp-demo' and base.parent != base:
    base = base.parent

df = pd.read_csv(base / 'datasets' / 'ventas_tienda.csv')
df['total_bruto'] = df['unidades'] * df['precio_unitario']
df['total_neto'] = df['total_bruto'] * (1 - df['descuento_pct'])
```

## Total neto por sucursal

```python
resumen_sucursal = df.groupby('sucursal', as_index=False)['total_neto'].sum().sort_values('total_neto', ascending=False)
resumen_sucursal.plot(x='sucursal', y='total_neto', kind='bar', title='Total neto por sucursal')
plt.ylabel('Total neto')
plt.tight_layout()
```

## Total neto por categoría

```python
resumen_categoria = df.groupby('categoria', as_index=False)['total_neto'].sum().sort_values('total_neto', ascending=False)
resumen_categoria.plot(x='categoria', y='total_neto', kind='bar', title='Total neto por categoría')
plt.ylabel('Total neto')
plt.tight_layout()
```

## Interpretación

Describe dos hallazgos a partir de los gráficos.

### Ejemplo de interpretación

- La sucursal con mayor total neto concentra una parte importante de las ventas.
- Una categoría domina el ingreso total, por lo que podría ser estratégica para promoción o stock.
