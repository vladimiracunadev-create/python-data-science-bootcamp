## Crear una métrica de apoyo

```python
# Qué hace: ejecuta el paso actual del ejercicio o laboratorio.
# Para qué sirve: deja explícita la intención del bloque y ayuda a revisarlo con más criterio.

df['total_bruto'] = df['unidades'] * df['precio_unitario']
df['total_neto'] = df['total_bruto'] * (1 - df['descuento_pct'])
df[['producto', 'total_neto']].head()
```

## Resumen por sucursal

```python
# Qué hace: resume la informaci?n por categoría o entidad.
# Para qué sirve: permite comparar grupos con una métrica agregada y no fila por fila.

df.groupby('sucursal', as_index=False)['total_neto'].sum().sort_values('total_neto', ascending=False)
```
