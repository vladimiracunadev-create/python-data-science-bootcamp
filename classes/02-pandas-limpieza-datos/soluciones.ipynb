# Clase 02 — Introducción a pandas

Usaremos `ventas_tienda.csv` para explorar datos de manera práctica.

```python
import pandas as pd
from pathlib import Path

base = Path.cwd()
while base.name != 'python-data-science-bootcamp-demo' and base.parent != base:
    base = base.parent

df = pd.read_csv(base / 'datasets' / 'ventas_tienda.csv')
df.head()
```

## Estructura del dataset

```python
df.info()
```

```python
df.describe(include='all')
```

## Selección de columnas

```python
df[['sucursal', 'categoria', 'unidades']].head()
```

## Filtros

```python
df[df['sucursal'] == 'Centro']
```

## Crear una métrica de apoyo

```python
df['total_bruto'] = df['unidades'] * df['precio_unitario']
df['total_neto'] = df['total_bruto'] * (1 - df['descuento_pct'])
df[['producto', 'total_neto']].head()
```

## Resumen por sucursal

```python
df.groupby('sucursal', as_index=False)['total_neto'].sum().sort_values('total_neto', ascending=False)
```

## Mini desafío

Responde: ¿qué categoría tiene mayor total neto acumulado?

```python
df.groupby('categoria', as_index=False)['total_neto'].sum().sort_values('total_neto', ascending=False)
```
