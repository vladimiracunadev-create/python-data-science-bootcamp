## Solución — Ejercicio 3: CRISP-DM para una tienda en verano

**Problema elegido:** Una tienda quiere saber qué productos tienen más probabilidad de venderse en verano.

| Paso CRISP-DM | ¿Qué haríamos? |
|---|---|
| Entender el problema | Definir "productos de verano" y qué significa "más probabilidad" (umbral de ventas) |
| Recolectar y entender los datos | Reunir ventas históricas de los últimos 3 veranos por producto y categoría |
| Preparar los datos | Filtrar por meses de verano, calcular promedios de venta por producto, manejar faltantes |
| Modelar | Ordenar productos por volumen promedio en verano; opcionalmente usar clustering |
| Evaluar | Comparar predicción con ventas reales del próximo verano |
| Comunicar e implementar | Entregar lista de los 10 productos más probables para reabastecimiento anticipado |

```python
# Solución — Ejercicio 4 (código de apoyo): estadísticas que responden preguntas descriptivas
# Qué hace: demuestra cómo los cuatro tipos de preguntas se traducen en código diferente
# Para qué sirve: conectar la teoría con la práctica antes de entrar a clases más técnicas

import pandas as pd

df = pd.read_csv("../../datasets/ventas_tienda.csv")

# PREGUNTA DESCRIPTIVA: ¿cuánto se vendió en total?
# sum() suma todos los valores de una columna numérica
if 'total_venta' in df.columns:
    total = df['total_venta'].sum()
    print(f"Pregunta descriptiva — Total vendido: {total:,.2f}")

# PREGUNTA DESCRIPTIVA: ¿cuál es la categoría más vendida?
# value_counts() cuenta ocurrencias de cada valor único
if 'categoria' in df.columns:
    print("\nCategorías más frecuentes en las ventas:")
    print(df['categoria'].value_counts().head())

# PREGUNTA DIAGNÓSTICA (exploratoria): ¿qué región tiene el precio promedio más alto?
# groupby + mean() agrupa registros y calcula el promedio de cada grupo
if 'region' in df.columns and 'precio_unitario' in df.columns:
    print("\nPrecio promedio por región:")
    print(df.groupby('region')['precio_unitario'].mean().sort_values(ascending=False))
```
