## ¿Qué es la ciencia de datos?

La ciencia de datos es la disciplina que extrae conocimiento útil de los datos para apoyar decisiones. Combina:

- **Programación** — para procesar millones de registros sin esfuerzo manual
- **Estadística** — para encontrar patrones reales y no confundirlos con ruido
- **Dominio del problema** — para saber qué pregunta tiene sentido hacer

Sin los tres ingredientes, el resultado puede ser incorrecto, inútil, o ambos.

## El ciclo de vida de un proyecto de datos

Todo proyecto de ciencia de datos sigue pasos similares. El estándar más conocido se llama **CRISP-DM**:

1. **Entender el problema** — ¿Qué decisión quiero tomar?
2. **Recolectar y entender los datos** — ¿Qué datos tengo? ¿Son confiables?
3. **Preparar los datos** — Limpiar, transformar, unir tablas
4. **Modelar** — Aplicar técnicas estadísticas o de machine learning
5. **Evaluar** — ¿El resultado responde la pregunta original?
6. **Comunicar e implementar** — Presentar hallazgos, tomar decisiones

Este bootcamp cubre cada uno de estos pasos. En las clases anteriores ya trabajaste varios de ellos sin saberlo.

## Tipos de datos en Python

Antes de explorar un dataset, recordemos qué tipos de datos maneja Python nativamente.

En ciencia de datos encontramos:
- **Numéricos**: precios, temperaturas, edades
- **Texto**: comentarios, nombres, categorías
- **Booleanos**: verdadero/falso (¿el cliente pagó?)
- **Listas/Secuencias**: múltiples valores juntos
- **Series de tiempo**: valores ordenados por fecha

```python
# Qué hace: muestra los tipos de datos nativos de Python que usamos en ciencia de datos
# Para qué sirve: entender que todo dato tiene un tipo, y el tipo determina qué operaciones podemos hacer

# Dato numérico entero
unidades_vendidas = 150
print(f"Unidades vendidas: {unidades_vendidas} — tipo: {type(unidades_vendidas)}")

# Dato numérico decimal
precio_unitario = 29.99
print(f"Precio unitario: {precio_unitario} — tipo: {type(precio_unitario)}")

# Dato de texto (string)
nombre_producto = "Cuaderno A4"
print(f"Producto: {nombre_producto} — tipo: {type(nombre_producto)}")

# Dato booleano
pago_realizado = True
print(f"¿Pago realizado? {pago_realizado} — tipo: {type(pago_realizado)}")

# Lista de valores (como una columna de datos)
ventas_semana = [120, 95, 200, 180, 145, 210, 88]
print(f"Ventas por día: {ventas_semana} — tipo: {type(ventas_semana)}")
```

## Los cuatro tipos de preguntas de la ciencia de datos

No todas las preguntas son iguales. Dependiendo del tipo de pregunta, usaremos técnicas distintas:

| Tipo | Pregunta | Ejemplo |
|---|---|---|
| **Descriptiva** | ¿Qué pasó? | ¿Cuánto vendimos este mes? |
| **Diagnóstica** | ¿Por qué pasó? | ¿Por qué bajaron las ventas en diciembre? |
| **Predictiva** | ¿Qué pasará? | ¿Cuánto venderemos el próximo trimestre? |
| **Prescriptiva** | ¿Qué debemos hacer? | ¿A qué clientes ofrecemos descuentos? |

En este bootcamp trabajaremos principalmente preguntas descriptivas y predictivas.

```python
# Qué hace: calcula estadísticas básicas sobre una lista de ventas
# Para qué sirve: responder una pregunta DESCRIPTIVA — ¿cómo fueron las ventas esta semana?

ventas_semana = [120, 95, 200, 180, 145, 210, 88]
dias = ["Lun", "Mar", "Mié", "Jue", "Vie", "Sáb", "Dom"]

# Total de la semana
total = sum(ventas_semana)
print(f"Total ventas semana: {total} unidades")

# Promedio diario
promedio = total / len(ventas_semana)
print(f"Promedio diario: {promedio:.1f} unidades")

# Mejor y peor día
mejor_dia = dias[ventas_semana.index(max(ventas_semana))]
peor_dia = dias[ventas_semana.index(min(ventas_semana))]
print(f"Mejor día: {mejor_dia} ({max(ventas_semana)} unidades)")
print(f"Peor día: {peor_dia} ({min(ventas_semana)} unidades)")
```

## Ciencia de datos en la vida real

Estas son aplicaciones que probablemente ya usas sin saber que tienen ciencia de datos adentro:

- **Netflix / YouTube**: predicen qué video verás con más gusto según tu historial
- **Pronóstico del tiempo**: modelos que procesan millones de sensores atmosféricos
- **Google Maps**: estima el tiempo de llegada con datos de velocidad en tiempo real
- **Detección de fraude**: los bancos detectan patrones inusuales en tus compras
- **Filtros de spam**: clasifican correos como spam o no spam automáticamente

Todos estos sistemas siguen el mismo ciclo: recolectan datos → los procesan → aprenden patrones → hacen predicciones → mejoran con retroalimentación.
