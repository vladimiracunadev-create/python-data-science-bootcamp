# Clase 08 — Presentación de hallazgos con claridad

Objetivo: Transformar resultados técnicos en mensajes claros para audiencia no técnica.

```
hallazgos = ["Los clientes activos aumentan mes a mes.", "La pérdida mensual se mantiene bajo 8% en gran parte del período.", "El ingreso promedio sube de forma consistente."]
for item in hallazgos:
    print("-", item)
```
