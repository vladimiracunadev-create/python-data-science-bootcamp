## Recorrer datos con `for`

```python
# Qué hace: ejecuta el paso actual del ejercicio o laboratorio.
# Para qué sirve: deja explícita la intención del bloque y ayuda a revisarlo con más criterio.

ventas = [
    {'producto': 'Mouse', 'unidades': 2, 'precio_unitario': 8990},
    {'producto': 'Teclado', 'unidades': 1, 'precio_unitario': 15990},
    {'producto': 'Webcam', 'unidades': 3, 'precio_unitario': 25990},
]

for item in ventas:
    total_item = item['unidades'] * item['precio_unitario']
    print(item['producto'], '=>', total_item)
```

## Funciones

```python
# Qué hace: define una función reutilizable para encapsular una regla del ejercicio.
# Para qué sirve: evita repetir lógica y facilita explicar la intención del cálculo.

def calcular_total(unidades, precio_unitario, descuento_pct=0.0):
    total_bruto = unidades * precio_unitario
    total_neto = total_bruto * (1 - descuento_pct)
    return total_neto

print(calcular_total(3, 8990))
print(calcular_total(2, 15990, 0.10))
```
