Este notebook resume en pocos pasos la progresión: cargar, explorar, transformar y visualizar.

```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
base = Path.cwd()
while base.name != 'python-data-science-bootcamp-demo' and base.parent != base:
    base = base.parent

df = pd.read_csv(base / 'datasets' / 'ventas_tienda.csv')
df.head()
```

```python
df['total_neto'] = (df['unidades'] * df['precio_unitario']) * (1 - df['descuento_pct'])
resumen = df.groupby('sucursal', as_index=False)['total_neto'].sum().sort_values('total_neto', ascending=False)
resumen
```

```python
resumen.plot(x='sucursal', y='total_neto', kind='bar', title='Demo rápida: total neto por sucursal')
plt.tight_layout()
```
